```python
def transform_price(price):
    return int(reduce(lambda x, y: str(x) + y, price.split(' ')[:-1]))

df_dict = {
            'url': [],
            'title': [],
            'price': [],
            'name': [],
            'location': [],
            'created_at': [],
            'sold_at': [],
            'raw_text': []
        }
df_len = len(urls)
for i, ad in enumerate(urls):
    rsss = rss_data[rss_data['url'] == ad]
    htmls = html_data[html_data['url'] == ad]
    df = pd.concat([rsss, htmls])
#     df_len = len(df)
    created = df['created'].min()
    deleted = df[df['deleted'].notna()]['deleted'].max()
    try:
        soup = BeautifulSoup(htmls.iloc[0]['original_content'], "lxml")
    except:
        pass
    raw_text = re.sub(r'\r\n', '', soup.find('div', class_="popis").text)
    title = re.sub(r'\r\n', '', soup.find('h1', class_="nadpis").text)
    table = soup.find('td', class_="listadvlevo")
    rows = table.find_all('tr')
    params = {}
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        cols = [ele for ele in cols if ele]
        params[cols[0].strip(':')] = cols[1]
    price = params['Cena']
    name = params['Meno:Telefón']
    location = params['Lokalita']
    ad_params = {
        'url': ad,
        'title': title,
        'price': transform_price(price) if price.split(' ')[-1] == '€' else np.nan,
        'name': name,
        'location': location,
        'created_at': created,
        'sold_at': deleted,
        'raw_text': raw_text
    }
    for key, value in ad_params.items():
        df_dict[key].append(value)
    
    sys.stdout.write("\r" + str('Processing ad {}/{}'.format(i, df_len)))
    sys.stdout.flush()
#     print(ad_params)
#     print('')
#         print(df['created'].min(), df[df['deleted'].notna()]['deleted'].max())
```

# Final function to extract all relevant data

```python
def create_documents(path_to_df:str='crawled_ads.csv/crawled_ads.csv'):
    """Yields dictionaries in the form of document to be added to the database
    
    Args:
        path_to_df (str): path to dataframe
        
    Yields:
        dict: Document to be added to DB
    
    """
    def transform_price(price):
        """Transforms the price from string to integer
        
        Args:
            price (str): string to be transformed
            
        Returns:
            int: transformed price
            
        Example: string '10 000 €' gets transformed to int 10000
        """
        return int(reduce(lambda x, y: str(x) + y, price.split(' ')[:-1]))

#     Read the dataframe
    data = pd.read_csv(path_to_df, sep=';', index_col=0, parse_dates=['created', 'modified', 'deleted'])
    urls = data['url'].unique()
    df_len = len(urls)
    for i, ad in enumerate(urls):
#         Prepare a document structure
        document = {
            'VIN': '',
            'license_plate': '',
            'price': '',
            'first_created': '',
            'last_modified': '',
            'deleted': '',
            'features': {
                
            },
            'ads': []
        }
#     create a df containing only relevant ads
        df = data[data['url'] == ad]
#     create a df containing onnly ads of type 'html'
        htmls = df[df['type'] == 'html']
    
        created = df['created'].min()
        deleted = df[df['deleted'].notna()]['deleted'].max()
        last_mod = df['modified'].max()
        
        document['first_created'] = created
        document['deleted'] = deleted
        document['last_modified'] = last_mod
#         add each of the selected ads to the document
        for ad_num in range(len(htmls)):
#         parse the html
            current_ad = htmls.iloc[ad_num]
            created_at = current_ad['created']
            modified_at = current_ad['modified']
            sold_at = current_ad['deleted']
            try:
                soup = BeautifulSoup(current_ad['original_content'], "lxml")
            except:
                pass
#             get the fields we are looking for
            raw_text = re.sub(r'\r\n\t', '', soup.find('div', class_="popis").text)
            title = re.sub(r'\r\n\t', '', soup.find('h1', class_="nadpis").text)
            table = soup.find('td', class_="listadvlevo")
            rows = table.find_all('tr')
            params = {}
            for row in rows:
                cols = row.find_all('td')
                cols = [ele.text.strip() for ele in cols]
                cols = [ele for ele in cols if ele]
                params[cols[0].strip(':')] = cols[1]
            price = params['Cena']
            name = params['Meno:Telefón']
            location = params['Lokalita']
            ad_params = {
                'url': ad,
                'title': title,
                'price': transform_price(price) if price.split(' ')[-1] == '€' else np.nan,
                'name': name,
                'location': location,
                'created_at': created_at,
                'modified_at': modified_at,
                'sold_at': sold_at,
                'raw_text': raw_text
            }
            document['ads'].append(ad_params)
        yield document
```
